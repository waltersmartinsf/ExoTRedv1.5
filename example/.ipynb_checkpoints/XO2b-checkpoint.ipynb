# Information about the data used in this example:

Reduction and Calibration of the exoplanet XO2b observation in the filter B Harris.
Data from Zellem et al.(2015): http://arxiv.org/abs/1505.01063

# WARNING

You have to put your login.cl in the directory where you would like to run this script or a similar.

# (1) Load the package

````python
import exotred #import the package
````

# (2) Importing the data Information from the YAML file input

````python
print 'Obtain the information in the YAML file:'
data_path, save_path, input_file = exotred.input_info('../input/')
````

# (3) Creating the superbias image

````python
print 'Create the superbias image.'
exotred.masterbias(input_file) #create a superbias image
````

# (4) Creating the superflat image

````python
print 'Create the superflat image'
exotred.masterflat(input_file) #create a superflat image
````

# (5) Reduce the science images using superbias and superflat images

````python
print 'Reduce and calibrate science images'
exotred.science_reduction(input_file) #reduce and calibrate in flat e bias the science images
````

# (6) Obtain time information using astropy

````python
print 'Obtain time information of each science image'
exotred.time_info(input_file) #obtain the time information for each science image (based on astropy)
````

# (7) Obtain Airmass and time information using pyraf (default time info for next steps)

````python
print 'Obtain airmass and put the time information in the header of each science image'
#obtain the airmass, and put all information of time in the header (based on pyraf).
exotred.time_calibration(input_file) 
````
